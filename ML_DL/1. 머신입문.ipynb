# Linear regression

## 기존의 프로그램 방식

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#섭씨 온도를 받아 화씨 온도로 변환
def celcius_to_fahrenheit(x):
    return x*1.8 + 32
```

```python
input_c = int(input("섭씨: "))
print("화씨:",celcius_to_fahrenheit(input_c))
```

## 머신러닝 방식
- 1. 데이터 확보 및 생성
- 2. 데이터 전처리(훈련 데이터, 검증 데이터, 시험 데이터)
- 3. 모델 구성
- 4. 모델 학습 과정 설정
- 5. 모델 학습시키기
- 6. 모델 평가(시험 데이터 셋으로)
- 7. 모델 사용(입력값을 예측)

```python
#1~2 데이터
data_c = np.array(range(100))
```

```python
data_f = celcius_to_fahrenheit(data_c)#타겟 데이터
```

```python
from tensorflow.keras.models import Sequential# 모델 생성
from tensorflow.keras.layers import Dense #입력과 출력
```

```python
#3. 모델 구성
model = Sequential()
model.add(Dense(1, input_dim=1))#독립변수 1개, 타겟 데이터 1개
```

```python
#4. 모델 학습 과정
model.compile(loss="mse", optimizer="rmsprop", metrics=['mae'])
```

```python
# 학습 전 예측
print(model.predict([0]))
```

```python
# 모델 저장
model.save("model/before_learning.h5")
```

```python
#컴퓨터를 학습시키기 위해 normalize (전체의 편차 줄이기, 이상적인 편차 0~1)
scaled_data_C = data_c/100.0
scaled_data_F = data_f/100.0
print(scaled_data_C)
print()
print(scaled_data_F)
```

```python
#학습시키기
hist = model.fit(scaled_data_C, scaled_data_F, epochs=1000, verbose=2)#학습횟수
```

```python
plt.plot(hist.history['loss']) #loss 감소
plt.show()
```

```python
model.predict([0.01])
```

```python
model.save("model/after_learning.h5")
```

## 노이즈가 있는 데이터

```python
#평균이 0이고 표준편차가 0.1인 데이터 100
np.random.seed(703)
noise = np.array(np.random.normal(0,0.1,100))
noised_scaled_data_F = scaled_data_F + noise
```

```python
plt.scatter(x=scaled_data_C, y=noised_scaled_data_F)
plt.show()
```

```python
#3. 모델 구성
model2 = Sequential()
model2.add(Dense(1, input_dim=1))
#4. 학습과정
model2.compile(loss="mse", optimizer="rmsprop", metrics=["mae"])
print(model2.summary())
```

```python
#5. 모델 학습시키기
hist2 = model2.fit(scaled_data_C, noised_scaled_data_F, epochs=1000, verbose=2)
```

```python
plt.plot(hist2.history['loss']) #loss 감소
plt.show()
```

```python
print(model2.predict(np.array([0.01])))
```

```python
model2.save('model/noise_after_learning.h5')
```
