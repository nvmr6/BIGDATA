```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow.keras.utils as utils # 원핫 인코딩
from tensorflow.keras.models import Sequential #모델 생성시
from tensorflow.keras.layers import Dense, Activation #입력과 출력
```

# 데이터 셋 생성

```python
dataset = np.loadtxt('data/pima-indians-diabetes.csv',delimiter=',')
dataset.shape
```

```python
#학습용 데이터
x_train = dataset[:700,:8]
y_train = dataset[:700,8]
#테스트 데이터
x_test = dataset[700:,:8]
y_test = dataset[700:,8]
```

# 모델 구성하기

```python
model = Sequential()
#입력 8 출력 1 sigmoid
model.add(Dense(12, input_dim=8, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

# 학습과정 설정

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

# 학습시키기

```python
hist = model.fit(x_train, y_train, epochs=1500, batch_size=64)
```

# 모델 평가

```python
score = model.evaluate(x_test, y_test)
```
