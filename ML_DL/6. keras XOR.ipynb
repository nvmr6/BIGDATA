# XOR

```python
import numpy as np
from tensorflow.keras.models import Sequential #모델 객체 생성시생성시
from tensorflow.keras.layers import Dense #layer 생성시
import matplotlib.pyplot as plt
```

```python
#1. training data set
x_data = [[0,0], [0,1], [1,0], [1,1]]
y_data = [[0],   [1],   [1],   [0]]

#2. 모델 구성
model = Sequential()
model.add(Dense(10,input_dim=2, activation="relu"))
# W1 = tf.Variable(tf.random_normal([2,10]), name="weight1")
# b1 = tf.Variable(tf.random_normal([10]), name="bias1")
# layer1 = tf.nn.relu(tf.matmul(X, W1) + b1)
model.add(Dense(20, activation="relu"))# input_dim 자동
model.add(Dense(10, activation="relu"))
model.add(Dense(1, activation="relu"))
model.summary()

#3. 모델 학습과정
model.compile(loss="mse", optimizer="adam", metrics=['binary_accuracy'])
#cost = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=logits,labels=Y))

#4. 학습하기
fit_hist = model.fit(x_data, y_data, epochs=200, verbose=1)
```

```python
fit_hist.history.keys()
```

```python
for acc in fit_hist.history["binary_accuracy"]:
    print(acc, end='\t')
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
while True:
    input_list = input("0 또는 1 2개 입력(종료: 9): ").strip().split(" ")
    input_data = list(map(int, input_list))
    print(input_list)
    if input_data[0]==9:
        break;
    input_data = np.array(input_data).reshape(1,2)
    print("입력: ", input_data)
    print("예측: ",int(model.predict(input_data).round()))
```

# OR gate

```python
#1. training data set
x_data = [[0,0], [0,1], [1,0], [1,1]]
y_data = [[0],   [1],   [1],   [1]]

#2. 모델 구성
model = Sequential()
model.add(Dense(32,input_dim=2, activation="relu"))
model.add(Dense(1, activation="relu"))
model.summary()

#3. 모델 학습과정
model.compile(loss="mse", optimizer="adam", metrics=['binary_accuracy'])

#4. 학습하기
fit_hist = model.fit(x_data, y_data, epochs=300)

for acc in fit_hist.history["binary_accuracy"]:
    print(acc, end='\t')
plt.plot(fit_hist.history['loss'])
plt.show()
```
