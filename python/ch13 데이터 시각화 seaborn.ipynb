## 관련 함수

### relational plot: 관계형 그래프

```python
sns.scatterplot(x="petal_length",y="petal_width",data=iris,hue="species")
plt.show()
```

```python
sns.scatterplot(x="petal_length",y="petal_width",data=iris,hue="species")
sns.lineplot(x="petal_length",y="petal_width",data=iris)
```

```python
fig,axes = plt.subplots(ncols=2, figsize=(8,5))
plt.subplots_adjust(wspace=0.3)
sns.scatterplot(x="petal_length",y="petal_width",data=iris,hue="species", ax=axes[0])
sns.lineplot(x="petal_length",y="petal_width",data=iris,hue="species", ax=axes[1])
```

```python
#relplot: lineplot을 서브 플롯으로 표현
sns.relplot(x="petal_length",y="petal_width",data=iris,hue="species",kind="line",col="species",col_wrap=2)
```

### categorical plot: 범주형 그래프

```python
#x축이 범주형일 경우 scatter 부적합
ax = sns.scatterplot(x="species",y="petal_length",data=iris)
```

```python
#산점도 1
ax1 = sns.stripplot(x="species",y="petal_length",data=iris)
```

```python
#산점도 2
ax2 = sns.swarmplot(x="species",y="petal_length",data=iris)
```

```python
titanic.head()
```

```python
#성별에 따른 생존자
a = sns.barplot(x='sex', y='survived', data=titanic)
```

```python
#클래스별 생존자
a = sns.barplot(x='sex', y='survived', data=titanic, hue="class")
```

```python
#서브 플롯으로
a = sns.catplot(x='sex', y='survived', data=titanic, hue="class", ci=None, kind="bar", col="class")
```

```python
#deck별 관측수를 막대그래프로
sns.countplot(x="deck", data=titanic)
```

```python
#class 별 survived를 sex 별로 pointplot
sns.pointplot(x="class", y="survived", hue="sex", data=titanic)
```

```python
sns.boxplot(x="sex", y="age", data=titanic, hue="survived")
sns.violinplot(x="sex", y="age", data=titanic, hue="survived")
```

```python
sns.boxenplot(x="alive", y="age", data=titanic, hue="adult_male")
```
